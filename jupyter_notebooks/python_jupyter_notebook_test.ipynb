````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,2*np.pi,0.25)
y = np.sin(x)

# this is a comment
# now we are going to plot

plt.figure(1, # figure handle number
           figsize=(8,4) # # width and height
          )
plt.plot(x, y,
         linestyle='--',
         color='green',
         marker='<')
plt.title('My first plot',
          fontsize=32.25)


plt.xlabel('x axis',
           fontsize=16,
           color=(34/255,139/255,34/255, 0.5)) # RBGA color, forrest green 
                                               #alpha means see through
plt.ylabel('y axis',
           fontsize=23,
           color='#FF6347') #hex color, red called "tomato" 
plt.show()
````
